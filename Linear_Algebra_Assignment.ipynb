## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
red = [1, 1]

blue = np.multiply(5, red)
green = np.multiply(np.pi, red)
yellow = np.multiply(-np.exp(1), red)

plt.arrow(0,0, red[0], red[1],head_width=.05, head_length=0.05, color ='red', width = 0.05)
plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, green[0], green[1],head_width=.05, head_length=0.05, color ='green')
plt.arrow(0,0, yellow[0], yellow[1],head_width=.05, head_length=0.05, color ='yellow')

plt.xlim(-5,6)          
plt.ylim(-5,6)
plt.title("Scaled Vectors")
plt.show()

# The vectors are all in one line, just different lengths.
# The -e is in the same line but pointed in opposite direction
```

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{i} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{j} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```
i = np.array([1, 0, -1])
j = np.array([1, np.sqrt(2), 1])

np.dot(i, j)

# These two vectors are orthogonal because their dot product equals zero.
# This means that these two vectors are linearly independent
# No way to scale one to get the other -- on different planes
```
